# **Lists**
#Most versatile data structure type in python, items in a list can be heterogeneous types, they are mutable

```python
#Making a list
l1 = [1,3,4,5]
l2 = [2,4,5,6]
```

```python
#Concatenation
l3=l1+l2
print(l3)
```

```python
#Copying the list
l4=l1
print(l4)
```

```python
#Append function - add a new value to the list
l1.append(13)
print(l1)
```

```python
#extend a list
l2.extend(l1)
print(l2)
```

```python
#Removing an element
l2.remove(5)
print(l2)
```

```python
#Removing an element from the end and returning it
l2.pop()
print(l2)
```

```python
#Sorting a list
l2.sort()
print(l2)
```

```python
l2.pop(1)
```

```python
#Finding out index of an element
l2.index(6)
```

```python
#Finding how many times an element appears in the list
l2.count(2)
```

```python
#Function to remove all occurences of an element from a list
l5=[1,1,1,2,2]
l5=list(filter((1).__ne__, l5))
print(l5)
```

# **Dictionaries**
#Unordered set of key and values, they are mutable, values of keys are unique

```python
#Dictionary
dc1={}
dc1["Sam"]=90
dc1["Ram"]=80
dc1["Lam"]=70
print(dc1)
```

```python
#Nested dictionaries
dc2={"test2":{"Sam":90, "Reena":80}, "test3":{"Sam":62}}
print(dc2)
```

```python
#Creating dictionaries from arbitrary key and value expressions
dc3={x:x*3 for x in (3,4,9)}
print(dc3)
```

```python
d4=dict([('x',62),('y',42)])
print(d4)
```

# **Tuples**
#sequence data type, they are immutable, can contain heterogeneous elements

```python
#Tuples
data=(189, 20, 9090)
data[1]
```

```python
#simultaneous assignment in tuples
(name, rollno, dept, x)= ("Akanksha", 8, "Computational biology", [1,2,3])
print(name)
print(dept, x)
```

```python
#We can assign a value to a name
x=(0.2,0.9)
y=("a","b")
print(x,y)
```

```python
#nested tuples
z=x,y,(1,2,3)
print(z)
```

```python
print(z[1])
len(z)
```

```python
#Extracting position in tuples using slices
new=z[1:]
print(new)
```

# **Sets**
#Unordered collection with no duplicate elements, cannot be sliced or indexed, mathematical operations can be performed - union, intersection, etc.

```python
#sets
color ={"yellow", "white", "pink"}
print(color)
```

```python
#converting list to set
set(list1)
```

```python
garment = {"jeans", "sweater"}
#union
color|garment
```

```python
set1={1,2,3,4,5}
set2={2,4,6,8,10}
#intersection
set1&set2
```

```python
#complement or difference
set1-set2
```

```python
#Exclusive or
set1^set2
```

# **Strings**

```python
#strings
s="Python\nA programming language"
print(s)
#make another string consisting on 2 strings
t=s+"\nYou are studying Python"
print(t)
```
