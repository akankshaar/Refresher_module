# Numpy

```python
#Constructing 1D and 2D arrays
import numpy as np
#1D array
x = np.array([1, 2, 3, 4])
print(x.ndim)
print(x.shape)
#2D array
y = np.array([[1, 2, 3, 4],[5, 6, 7, 8]])
print(y.ndim)
print(y.shape)
```

```python
#List to array
l = [[1,2,3],[3,6,9],[2,4,6]] #create a list
a = np.array(l)
print(a)
print(a.shape)
print(a.dtype)
```

```python
#Operations on arrays
a=np.array([1,3,5,7,9])
b=np.array([3,5,6,7,9])
c=a+b
print(c)
```

```python
#Reshape a 1D array to 2D array
d=np.array([1,2,3,4,5,6])
print(d.ndim)
d.ravel()
d.reshape(2,3)
```

```python
#Reshape a 1D array to 3D array
d=np.array([1,2,3,4,5,6])
print(d.ndim)
d.ravel()
e = d.reshape(3,2,1)
print(e.ndim)
print(e.shape)
```

```python
#Find out data type, size occupied by array, and size of array
print(a.dtype.name)
print(a.itemsize)
print(a.size)
```

```python
#Appoint a datatype to an array
e = np.array([[1, 2], [3, 4]], dtype=complex)
print(e)
```

```python
#Make an array
a1 = np.arange(15)
a2 = np.arange(10, 30, 5)
a3 = np.zeros((3, 4))
print(a1)
print(a2)
print(a3)
```

```python
#Transpose an array
d.ravel()
f=d.reshape(2,3)
print(f)
np.transpose(f)
```

```python
#Sort an array
np.sort(c)
```

```python
#Sort an array using indices
np.argsort(c)
```

# Seaborn


```python
# importing packages 
import seaborn as sns 
```

```python
# loading dataset 
data = sns.load_dataset("iris") 
  
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
```

```python
# Python program to illustrate
# plotting using Swarmplot

ax = sns.swarmplot(x='species', y='sepal_length', data=data)
 
# giving title to the plot
plt.title('Graph')
 
# function to show plot
plt.show()
```

```python
#multiple plots
# loading of a dataframe from seaborn
df = sns.load_dataset('tips')
# Form a facetgrid using columns with a hue
graph = sns.FacetGrid(df, col ="sex",  hue ="day")
# map the above form facetgrid with some attributes
graph.map(plt.scatter, "total_bill", "tip", edgecolor ="w").add_legend()
# show the object
plt.show()
```

```python
# set grid style
sns.set(style ="darkgrid")

sns.relplot(x ="total_bill", y ="tip", 
            hue ="day", data = df);
```
